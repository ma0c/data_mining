```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(
    "prec_dia.csv", 
    delimiter=";",
    decimal=","
)
data
```

```python
just_data = data.iloc[:,3:]
just_data
```

```python
just_data.describe(include="all")
```

```python
%matplotlib inline
just_data.boxplot(rot=90,figsize=(12,8))
```

```python
just_data.isna().sum().plot(kind="bar", title="Missing Values",figsize=(12,8))
```

```python
Q1_Q3 = just_data.quantile([.25,.75])

Q1_Q3.loc["IQR"] = Q1_Q3.loc[0.75] - Q1_Q3.loc[0.25]
Q1_Q3.loc["H"] = Q1_Q3.loc["IQR"] * 1.5
Q1_Q3.loc["H_MAX"] = Q1_Q3.loc[0.75] + Q1_Q3.loc["H"]
Q1_Q3.loc["H_MIN"] = Q1_Q3.loc[0.25] - Q1_Q3.loc["H"]
Q1_Q3
```

```python
without_outliers = just_data[(just_data < Q1_Q3.loc["H_MAX"]) & (just_data > Q1_Q3.loc["H_MIN"])]
without_outliers
```

```python
without_outliers.boxplot(rot=90,figsize=(12,8))
```
